```python
# import packages
import pandas as pd
import numpy as np
import keras as keras
from scipy.stats import norm

# to plot within notebook
import matplotlib.pyplot as plt
%matplotlib inline

# read the file, Log Returns
df_log = pd.read_csv('Log_Industry_Return_Data.csv')
df_log.head(1)
```

```python
# Number of days and simulations
t_intervals = 1825
iterations = 10

# Imagine closing price of â‚¬100
inital = 100
```

```python
# Agric is column index 2
df_log.iloc[0,2]
```

```python
# Agric
ag_df = df_log.iloc[:,2]
u_ag = ag_df.mean()
v_ag = ag_df.var()
drift_ag = np.array(u_ag - (0.5 * v_ag))
std_ag = np.array(ag_df.std())

# Generate ag_returns 
ag_returns = np.exp(drift_ag + std_ag * norm.ppf(np.random.rand(t_intervals, iterations)))

# Find prices
ag_price_list = np.zeros_like(ag_returns)
ag_price_list[0] = inital

for t in range(1, t_intervals):
    ag_price_list[t] = ag_price_list[t - 1] * ag_returns[t]

# plot results
plt.plot(ag_price_list)

# turn to dataframe then to csv
ag_simulatedf = pd.DataFrame(ag_price_list)
ag_simulatedf.to_excel(r"C:\Users\jacka\Desktop\College Final Year\Semester 1\STAT 40920\Group Project\Simulations\Agric_Simulation.xlsx", 
               index=False)
```
